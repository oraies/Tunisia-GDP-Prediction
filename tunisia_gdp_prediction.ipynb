# Overview

This Jupyter Notebook demonstrates a simple data analysis and prediction task using GDP (Gross Domestic Product) data for Tunisia. The notebook is structured as follows:

1. **Data Processing**:
   - The GDP data from the file "gdp_1960_2020.csv" is loaded into a Pandas DataFrame.
   - Data specific to Tunisia is extracted from the DataFrame.

2. **Data Visualization**:
   - A bar plot is created to visualize the GDP trends for Tunisia over the years from 1965 to 2020.

3. **Model Training**:
   - A Linear Regression model is trained using the extracted data to predict Tunisia's GDP based on the years.

4. **GDP Prediction**:
   - The trained model is used to predict Tunisia's GDP for the years 2024 to 2030.
   - The predicted GDP values are printed out.


```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# Data Processing
global_df = pd.read_csv('./gdp_1960_2020.csv')
tunisia_df = global_df[global_df['country'] == 'Tunisia']

# Extract Data
y_train = tunisia_df['gdp']
x_train = tunisia_df.drop(columns=['state', 'gdp_percent', 'rank', 'country', 'gdp'], axis=1)  

# Plotting GDP over years
plt.figure(figsize=(10, 6))
plt.bar(x_train['year'], y_train, color='skyblue')
plt.title("Tunisia GDP 1960-2020")
plt.xlabel('Years')
plt.ylabel('GDP')
plt.show()
```

```python
# Model Training
model_tunisia = LinearRegression()
model_tunisia.fit(x_train, y_train)

# Plotting Actual vs Predicted GDP
pred = model_tunisia.predict(x_train)

plt.figure(figsize=(10, 6))
plt.scatter(x_train, y_train, label='Actual value')
plt.plot(x_train, pred, color='orange', label='Predicted value')
plt.legend()
plt.title("Tunisia GDP Prediction")
plt.xlabel("Years")
plt.ylabel("GDP")
plt.show()
```

```python
# Predicting GDP for future years
years = list(range(2024, 2031))
years_df = pd.DataFrame(years, columns=['year'])
predicted_gdp = model_tunisia.predict(years_df)

# Printing Predicted GDP for future years
for i, year in enumerate(years):
    print(f"Year: {year}, Predicted GDP: {predicted_gdp[i]}")
```
